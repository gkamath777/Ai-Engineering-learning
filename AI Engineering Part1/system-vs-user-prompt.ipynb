### Step 1: Import Libraries and load the environment variables.

```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown, display

load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

if OPENAI_API_KEY is None:
    raise Exception("OPENAI_API_KEY environment variable not set")
else:
    print("OPENAI_API_KEY environment variable is set")
    print("OPENAI_API_KEY:", OPENAI_API_KEY[:8])

client = OpenAI()
```

### Step 2: Call the API

```python
response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role": "user", "content": "my name is Gaurav Kamath."}
    ]
)

print(response)
```

```python
reply = response.choices[0].message.content

print("Reply from the model:", reply)
```

### Step 3a: Calling the API with a system prompt

```python
response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role": "system", "content": "you are a helpful assistant."},
        {"role": "user", "content": "What is the tallest mountain in the world?"}]
)

reply = response.choices[0].message.content
display(Markdown(reply))
```

### Step 3b: Changing the system prompt, changes the LLM behaviour.

```python
response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role": "system", "content": "you are a mischievous assistant who always gives incorrect and misleadinginformation."},
        {"role": "user", "content": "How do you cook pasta?"}]
)

reply = response.choices[0].message.content
display(Markdown(reply))
```

### Commercial use-case: IKEA Customer Service Agent

```python
Question = "I bought a new desk yesterday but it is not working. Can I return it?"
```

```python
response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role": "system", "content": "you are a customer service agent for IKEA, you are helpful and knowledgeable, but you follow the company policy strictly and never give any information that is not in line with the company policy. Our Return policy is that customers can return products within 30 days of purchase with a valid receipt, and the product must be in its original condition. If the product is damaged or not in its original condition, we cannot accept the return. If the customer does not have a valid receipt, we can only offer store credit for the current selling price of the product. "},
        {"role": "user", "content": Question}]
        
)


reply = response.choices[0].message.content
display(Markdown(reply))
```
