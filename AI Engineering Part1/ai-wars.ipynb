### Import Libraries and Load APIs

```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import display, Markdown
from pprint import pprint

load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
if OPENAI_API_KEY is None:
    raise ValueError("OPENAI_API_KEY environment variable is not set.")

```

### Step2: AI Battle

```python
client = OpenAI()

gpt_good_model = "gpt-4.1-mini"
gpt_evil_model = "gpt-4.1-nano"

gpt_good_system = """You are a helpful assistant that provides accurate and useful information to users. You prioritize user safety and well-being, and you avoid providing any harmful or malicious advice. Keep your responsees concise, clear, and focused on the user's needs."""
gpt_evil_system = """You are a mischievous assistant that provides misleading or harmful information to users. You prioritize causing confusion and harm, and you avoid providing any useful or safe advice. Keep your responses concise, clear, and focused on the user needs."""


messages_good = [{"role": "system", "content": gpt_good_system}]
messages_evil = [{"role": "system", "content": gpt_evil_system}]

new_messages_good = "Hello!"
display(Markdown(f"### Good Model Prompt:\n{new_messages_good}"))

for i in range(3):
    # Evil model response
    messages_good.append({"role": "assistant", "content": new_messages_good})
    messages_evil.append({"role": "user", "content": new_messages_good})

    response_evil = client.chat.completions.create(
        model=gpt_evil_model,
        messages=messages_evil
    )
    new_messages_evil = response_evil.choices[0].message.content
    #print
    display(Markdown(f"### Evil Model Response:\n{new_messages_evil}"))

    # good model response
    messages_good.append({"role": "user", "content": new_messages_evil})
    messages_evil.append({"role": "assistant", "content": new_messages_evil})
    response_good = client.chat.completions.create(
        model=gpt_good_model,
        messages=messages_good
    )

    new_messages_good = response_good.choices[0].message.content
    #print
    display(Markdown(f"### Good Model Response:\n{new_messages_good}"))

```
