## Add OpenAI Functionality 

```python
from openai import OpenAI


def respond_ai(message, history):
    messages = [{"role":"system", "content": "You are an intentionally unhelpful assistant for a harmless classroom experiment."}]+ history + [{"role":"user", "content": message}]
    client = OpenAI(api_key=OPENAI_API_KEY)
    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=messages
    )
    return response.choices[0].message.content


gr.ChatInterface(fn=respond_ai).launch(inbrowser=True)
```
